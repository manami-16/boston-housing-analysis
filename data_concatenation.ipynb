```python
GBREB_df = pd.read_csv('data/pdf_generated.csv')
GBREB_df.head()
```

```python
selected_GBREB = ['Time', 'Units Sold', "Active Listings", "Months Supply of Inventory", 
                 "New Listings", "Pending Sales", "Days to Off Market", "Sold to Original Price Ratio", 
                 "Price per Square Foot", 'Median Selling Price']
```

```python
GBREB_df = GBREB_df[selected_GBREB]
GBREB_df.head()
```

```python
## convert time column into month and year
year = []
month = []

for i in range(len(GBREB_df['Time'])):
    frac, whole = math.modf(GBREB_df['Time'][i])
    year.append(round(whole))
    month.append(round(round(frac, 2) * 100))
```

```python
GBREB_df['year'] = year
GBREB_df['month'] = month
GBREB_df.drop(columns='Time', axis=1, inplace=True)
```

```python
GBREB_df.head()
```

```python
GBREB_df['MEDV'] = GBREB_df['Median Selling Price'] / 1000
GBREB_df.drop(columns=['Median Selling Price'], axis=1, inplace=True)
```
